```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

months= ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
received = [160, 184, 241, 149, 180, 161, 132, 202, 160, 139, 149, 177]
processed = [160 ,184 ,237 ,148 ,181 ,150 ,123 ,156 ,126 ,104 ,124 ,140]
```

```python
plt.plot(months, received,color='#00ff00',label='Recebidos',)
plt.plot(months, processed,color='#ff0000', label='Processados')
plt.ylim(90, 250)
plt.title('Processos Recebidos e Processados em cada mÃªs',fontdict={'fontsize': 15})
plt.legend()
plt.show()
```
